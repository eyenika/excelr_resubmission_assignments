```python
#preprocessing #checking missing values
print("\nMissing values before handling:")
print(zoo.isnull().sum())
```

```python
#handling missing values
df = zoo.fillna(zoo.mean(numeric_only=True))  # numerical columns
df = zoo.fillna(zoo.mode().iloc[0]) #categorical columns
```

```python
#seperate features and target
X = zoo.drop('animal name', axis=1)  # Changing 'animal_type' to 'animal name'
y = zoo['animal name']  # Changing 'animal_type' to 'animal name'
     
```

```python
#identify and handle non-numeric columns more thoroughly
non_numeric_cols = X.select_dtypes(include=['object', 'bool']).columns
print("\nNon-numeric columns:", non_numeric_cols)

for col in X.columns:
    # Checking if column contains boolean strings
    unique_vals = set(str(x).lower() for x in X[col].unique())
    if {'true', 'false'}.issubset(unique_vals):
        X[col] = X[col].replace({'true': 1, 'false': 0, 'True': 1, 'False': 0}, inplace=False)
        X[col] = pd.to_numeric(X[col], errors='coerce')
    # Converting actual boolean columns
    elif X[col].dtype == 'bool':
        X[col] = X[col].astype(int)
    # Converting other categorical columns with few unique values
    elif X[col].dtype == 'object' and X[col].nunique() < 10:
        X[col] = LabelEncoder().fit_transform(X[col])
    # Droping columns with many unique values (like names)
    elif X[col].dtype == 'object' and X[col].nunique() >= 10:
        X = X.drop(col, axis=1)
```

```python
# Ensuring all remained columns are numeric
X = X.apply(pd.to_numeric, errors='coerce')
X = X.dropna(axis=1)  # cannot be converted will be dropped
```

```python
print("\nColumns after converting:")
print(X.dtypes)
```

```python
# Handle outliers using Z-score
z_scores = zscore(X)
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
X = X[filtered_entries]
y = y[filtered_entries]

print("\nNumber of rows left removing outliers:", len(X))
```
