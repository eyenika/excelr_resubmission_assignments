```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_excel("Online retail.xlsx", header=None)
df
```

```python
#preprocessing
transactions = df[0].dropna().apply(lambda row: [item.strip() for item in row.split(',')]).tolist()
transactions
```

```python
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
basket_df = pd.DataFrame(te_array, columns=te.columns_)
```

```python
te
```

```python
te_array
```

```python
basket_df
```

```python
#frequent items using apropi algorithm
frequent_itemsets = apriori(basket_df, min_support=0.03, use_colnames=True)
```

```python
frequent_itemsets
```

```python
#association rules generation
from mlxtend.frequent_patterns import association_rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```

```python
rules
```

```python
print(frequent_itemsets)
```

```python
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
#What is lift and why is it important in association rules?
->Lift will tell how  more likely two items are bought together when compared to just by chance.
->if the value is more than 1 then we can say that the relationship is more strong.
```

```python
#What is support and confidence? How do you calculate them?
->support shows how often the group of items are seen together in the dataset.
->confidence shows how often an item is bought together when other item is already bought together.
```

```python
#What are some limitations or challenges of Association Rule Mining?
->does not show cause-and-effect, just patterns.
->can create too many rules, which can be overwhelming
```
